```python
import scanpy as sc
import pandas as pd

# Read the .h5ad file with scanpy
sciplex = sc.read_h5ad('/Mounts/rbg-storage1/users/johnyang/cellot/datasets/scrna-sciplex3/hvg.h5ad')
```

```python
print("Shape of sciplex.X: ", sciplex.X.shape)
print("Length of sciplex.var.index: ", len(sciplex.var.index))
print("Length of sciplex.obs.index: ", len(sciplex.obs.index))
```

```python
from scipy.sparse import csr_matrix

sparse_matrix = csr_matrix(sciplex.X)

# Then convert to a DataFrame.
df = pd.DataFrame.sparse.from_spmatrix(sparse_matrix, index=sciplex.obs.index, columns=sciplex.var.index)
```
